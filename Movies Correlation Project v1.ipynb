```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

# Now lets read the data into the notebook
df= pd.read_csv(r'C:\Users\Admin\Downloads\movies.csv')
```

```python
df = df.dropna(how='any',axis=0)

df
```

```python
# Lets check to see if there is any missing data
# We'll do this by looping over the entire columns in the data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, round(pct_missing * 100)))
```

```python
# Lets check for duplicates
df.drop_duplicates()
```

```python
#Now, Lets order the data
df.sort_values(by=['gross'], inplace= False, ascending=False)
```

```python
# Showing the correlation heatmap for the above data

correlation_matrix= df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Numeric Values')

plt.xlabel('Movie features')

plt.ylabel('Movie features')

plt.show()
```

```python
# Lets convert all values in the dataset to categorized values
# We'll do this using factorize

df.apply(lambda x:x.factorize()[0]).corr(method='pearson')
```

```python
# Lets have a heatmap for this correlation

correlation_matrix= df.apply(lambda x:x.factorize()[0]).corr(method='pearson')

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for all Values')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#Lets have a correlation pair

correlation_matrix= df.apply(lambda x: x.factorize()[0]).corr(method ='pearson')

corr_pairs = correlation_matrix.unstack()

print(corr_pairs)
```

```python
#Lets sort out the values

sorted_pairs= corr_pairs.sort_values(kind= "quicksort")

print(sorted_pairs)
```

```python
#Lets take a look at those with high correlation (strong_pairs > 0.5)

strong_pairs= sorted_pairs[abs(sorted_pairs) > 0.5][:50]
print(strong_pairs)
```

```python
#This assigns a numeric value to all features

df_numerized = df

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == "object"):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
df_numerized.corr(method= 'pearson')
```

```python
#Heatmap showing the correlation matrix for Movie features

correlation_matrix= df_numerized.corr()

sns.heatmap(correlation_matrix, annot= True)

plt.title('Correlation Matrix for Movie Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
